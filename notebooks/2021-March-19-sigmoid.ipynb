```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def sig(z):
    return 1/(1+np.exp(-z))
```

```python
sig(0)
```

```python
sns.set_context("talk")

x = np.linspace(-5, 5, 500)
plt.plot(x, sig(x),color='k')
plt.xlabel("z", fontsize=20)
plt.ylabel(r"$\sigma(z) = \frac{1}{1+e^{-z}}$", fontsize=22)
sns.despine()
plt.savefig("/Users/nipun/git/ml-slides/logistic-regression/sigmoid.pdf", dpi=400, transparent=True, bbox_inches="tight")
```
